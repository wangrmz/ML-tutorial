1）混淆矩阵
混淆矩阵（Confusion Matrix）是用于评估分类模型性能的工具，展示了模型预测结果与实际标签的对比情况。

```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
# 定义列表标签
label = ["猫", "狗"]  # 标签
# 定义数据 ： 定义预测值和真实值
y_true = ["猫", "猫", "猫", "猫", "猫", "猫", "狗", "狗", "狗", "狗"]  # 真实值
y_pred1 = ["猫", "猫", "狗", "猫", "猫", "猫", "猫", "猫", "狗", "狗"]  # 预测值
# 得到混淆矩阵
matrix1 = confusion_matrix(y_true, y_pred1, labels=label)  # 混淆矩阵
print(pd.DataFrame(matrix1, columns=label, index=label))
# 使用热力图
sns.heatmap(matrix1, annot=True, fmt='d', cmap='Greens')
```

```python
from sklearn.metrics import accuracy_score

# 准确率（Accuracy）
# 正确预测的比例。准确率
accuracy = accuracy_score(y_true, y_pred1)
print(accuracy)  # 0.7
```

```python
# 3）精确率（Precision）
# 预测为正例的样本中实际为正例的比例，也叫查准率。
from sklearn.metrics import precision_score

# 指明正类标签pos_label
precision = precision_score(y_true, y_pred1, pos_label="猫")
print(precision)  # 0.7142857142857143
```

```python
# 4）召回率（Recall）
# 实际为正类的样本中预测为正类的比例，也叫查全率。
# 预测为正的/实际上为正的（真正例+假负例）
from sklearn.metrics import recall_score

recall = recall_score(y_true, y_pred1, pos_label="猫")  # pos_label指定正例
print(recall)  # 0.8333333333333334
```

```python
# 5）F1分数（F1 Score）
# 精确率和召回率的调和平均。
from sklearn.metrics import f1_score

f1 = f1_score(y_true, y_pred1, pos_label="猫")  # pos_label指定正例
print(f1)  # 0.7692307692307693
```

```python
# 在代码中，我们可通过sklearn.metrics.classification_report生成分类任务的评估报告，包括精确率、召回率、F1分数等。
from sklearn.metrics import classification_report

report = classification_report(y_true, y_pred1, labels=label, target_names=None)
print(report)
# y_true：真实标签
# y_pred：预测的标签
# labels：可选，指定需要计算的类别列表（默认计算所有出现过的类别）
# target_names：可选，类别名称（默认使用 labels 指定的类别号）
'''
                 precision   recall    f1-score     support

           猫       0.71      0.83      0.77         6
           狗       0.67      0.50      0.57         4

    accuracy                           0.70        10
   macro avg       0.69      0.67      0.67        10
weighted avg       0.70      0.70      0.69        10
'''
```
