```python
# 构造特征
a  = np.random.randn(100)
print(a)
# 方差
print(np.var(a))
```

```python
# b = np.random.randn(100) * 0.1
'''
b = np.random.normal(5, 0.1, 100) 的意思是：从正态分布中随机抽取 100 个样本。
具体参数含义：
5：分布的均值（μ = 5），即数据的中心位置。
0.1：标准差（σ = 0.1），衡量数据围绕均值的离散程度。值越小，数据越集中。
100：生成的随机样本数量，返回一个长度为 100 的一维 NumPy 数组。
所以 b 中会包含 100 个大致以 5 为中心、波动幅度约 0.1 的随机浮点数。
'''
b = np.random.normal(5,0.1,100)
# print(b)
print(np.var(b)) # 0.009671820511906154
```

```python
# 构造特征向量（输入数据X）
# 进行转置 a，b 从100列 2行转成100行2列
X = np.vstack((a,b)).T
print(X)
print(X.shape) # (100, 2)
```

```python
# 低方差过滤
# 低方差 = 低信息量 = 可以扔掉，不会影响模型性能，反而能减少计算量、避免过拟合
from sklearn.feature_selection import VarianceThreshold
# 低于指定阈值的就直接过滤掉
# 删除方差低于0.01的特征
vt = VarianceThreshold(0.01)
X_filtered = vt.fit_transform(X)
print(X_filtered)



```

```python
from sklearn.feature_selection import VarianceThreshold
import numpy as np

# 生成两列数据
a = np.random.randn(100)          # 方差大约 1
b = np.random.normal(5, 0.1, 100) # 方差大约 0.01

X = np.vstack((a, b)).T

vt = VarianceThreshold(0.01)
X_filtered = vt.fit_transform(X)

print("原始形状:", X.shape)        # (100, 2)
print("过滤后形状:", X_filtered.shape) # 大概率 (100, 1)
print("哪些列被保留？", vt.get_support()) # 例如 [ True False ]
```

因为机器学习的数据格式是 行 = 样本，列 = 特征。我们关心的是：这个特征在所有样本上的取值变化大不大。如果它在不同样本间几乎没变化，就没用。

按列计算。每一列就是一个特征，VarianceThreshold 会分别计算每一列（即每一个特征）的方差，然后看它是否小于阈值（0.01），小于就删掉这一整列。


计算 列0 的方差：用 a 的 100 个数值，套用方差公式
Var(列0) = 平均( (每个值 - 均值)^2 ) → 得到一个数值（例如 1.02）

计算 列1 的方差：用 b 的 100 个数值 → 你之前打印过，np.var(b) ≈ 0.00967

比较：

列0 的方差 1.02 ≥ 0.01 → 保留列0

列1 的方差 0.00967 < 0.01 → 删除列1
